Model

UnderSampling

```python
credit_df.columns
```

```python
normal=credit_df[credit_df['Class']==0]
fraud=credit_df[credit_df['Class']==1]
```

```python
print(normal.shape,fraud.shape)
```

```python
normal_data=normal.sample(n=473)
```

```python
normal_data.shape
```

```python
credit_new_df=pd.concat([normal_data,fraud],ignore_index=True)
```

```python
credit_new_df.head()
```

```python
credit_new_df['Class'].value_counts()
```

```python
X=credit_new_df[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11',
       'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21',
       'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
y=credit_new_df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
classifier={
    "Logistic Regression": LogisticRegression(),
    "RandomForestClassifier": RandomForestClassifier()
}

for model,cls in classifier.items():
    print(f"\n------{model}------")
    cls.fit(X_train,Y_train)
    test_pred=cls.predict(X_test)
    print(f"\n Accuracy: {accuracy_score(Y_test,test_pred)}")
    print(f"\n Precision: {precision_score(Y_test,test_pred)}")
    print(f"\n Recall: {recall_score(Y_test,test_pred)}")
    print(f"\n F1 Score: {f1_score(Y_test,test_pred)}")
```

OverSampling

```python
X=credit_df[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11',
       'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21',
       'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
y=credit_df['Class']
```

```python
print(X.shape,y.shape)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res=SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_res,y_res,test_size=0.2,random_state=4)
```

```python
classifier={
    "Logistic Regression": LogisticRegression(),
    "RandomForestClassifier": RandomForestClassifier()
}

for model,cls in classifier.items():
    print(f"\n------{model}------")
    cls.fit(X_train,Y_train)
    test_pred=cls.predict(X_test)
    print(f"\n Accuracy: {accuracy_score(Y_test,test_pred)}")
    print(f"\n Precision: {precision_score(Y_test,test_pred)}")
    print(f"\n Recall: {recall_score(Y_test,test_pred)}")
    print(f"\n F1 Score: {f1_score(Y_test,test_pred)}")
```
